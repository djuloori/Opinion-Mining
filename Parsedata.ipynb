# Parsing the JSON Data

```python
import pandas as pd
import gzip

def parse(path):
  g = gzip.open('C:/Users/Sudha/Documents/GitHub/Opinion-Mining/reviews_Digital_Music_5.json.gz', 'rb')
  for l in g:
    yield eval(l)

def getDF(path):
  i = 0
  df = {}
  for d in parse(path):
    df[i] = d
    i += 1
  return pd.DataFrame.from_dict(df, orient='index')

Parsed_df = getDF('C:/Users/Sudha/Documents/GitHub/Opinion-Mining/reviews_Digital_Music_5.json.gz')
```

# Deleting Unwanted Columns

```python
Parsed_df = Parsed_df[["reviewText","overall"]].copy()
Parsed_df.head(100)
```

# Determining the Classes

```python
Parsed_df['rating'] = '1'
Parsed_df['rating'][Parsed_df['overall'] > 3] = 'Good'
Parsed_df['rating'][Parsed_df['overall'] <= 3] = 'Bad'
```

# Writing the data into a CSV file

```python
Parsed_df.to_csv('dataset.csv')
```
